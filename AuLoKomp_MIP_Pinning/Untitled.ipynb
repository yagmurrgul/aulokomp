```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the column names (in the order in the file)
col_names = [
    "Instance_Name", 
    "Completion_Time", 
    "Objective_Value", 
    "Gap", 
    "Nb_ULs", 
    "Nb_Desired_ULs", 
    "Grid_Occupancy", 
    "UL_Occupancy"
]
```

```python

# Read the file; skip the header line and use our own column names
df = pd.read_csv("combined_output.txt", delim_whitespace=True, names=col_names, skiprows=1)
```

```python
# Convert numeric columns to appropriate types
numeric_cols = ["Completion_Time", "Objective_Value", "Gap", "Nb_ULs", "Nb_Desired_ULs", "Grid_Occupancy", "UL_Occupancy"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```python

# Graph 1: Histogram of Completion Time
plt.figure(figsize=(8, 6))
plt.hist(df["Completion_Time"], bins=30, edgecolor='black')
plt.xlabel("Completion Time (ms)")
plt.ylabel("Frequency")
plt.title("Distribution of Completion Time")
plt.tight_layout()
plt.show()

# Graph 2: Scatter plot of Completion Time vs. Grid Occupancy
plt.figure(figsize=(8, 6))
plt.scatter(df["Grid_Occupancy"], df["Completion_Time"], alpha=0.7)
plt.xlabel("Grid Occupancy")
plt.ylabel("Completion Time (ms)")
plt.title("Completion Time vs. Grid Occupancy")
plt.tight_layout()
plt.show()

# Graph 3: Scatter plot of Objective Value vs. Completion Time
plt.figure(figsize=(8, 6))
plt.scatter(df["Completion_Time"], df["Objective_Value"], alpha=0.7)
plt.xlabel("Completion Time (ms)")
plt.ylabel("Objective Value")
plt.title("Objective Value vs. Completion Time")
plt.tight_layout()
plt.show()

# Graph 4: Box Plot of Completion Time grouped by Nb_ULs
# Convert Nb_ULs to a categorical variable (if there are only a few distinct values)
plt.figure(figsize=(10, 6))
unique_sizes = sorted(df["Nb_ULs"].unique())
data_to_plot = [df[df["Nb_ULs"] == size]["Completion_Time"] for size in unique_sizes]
plt.boxplot(data_to_plot, labels=unique_sizes)
plt.xlabel("Nb_ULs (System Size)")
plt.ylabel("Completion Time (ms)")
plt.title("Completion Time by Number of ULs")
plt.tight_layout()
plt.show()

# Graph 5: Correlation Heatmap for the numeric columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
im = plt.imshow(corr, cmap="viridis", interpolation="none")
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(ticks=np.arange(len(numeric_cols)), labels=numeric_cols, rotation=45, ha="right")
plt.yticks(ticks=np.arange(len(numeric_cols)), labels=numeric_cols)
plt.title("Correlation Matrix of Numerical Features")
plt.tight_layout()
plt.show()
```
